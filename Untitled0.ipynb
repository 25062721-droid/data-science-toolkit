```python
import numpy as np

# 1.1 Create a numpy array of size 2x2
array_2x2 = np.array([[1, 2], [3, 4]])
print("1. 2x2 Array:")
print(array_2x2)
print()

# 1.2 Create a random numpy array of size 3x4
random_3x4 = np.random.rand(3, 4)
print("2. 3x4 Random Array:")
print(random_3x4)
print()

# 1.3 Create a numpy array of ones and a numpy array of zeros
ones_array = np.ones((2, 3))
zeros_array = np.zeros((3, 2))
print("3. Ones Array (2x3):")
print(ones_array)
print("Zeros Array (3x2):")
print(zeros_array)
print()

# 1.4 Create a random numpy array of size 6x3 and transpose it
random_6x3 = np.random.rand(6, 3)
transposed_6x3 = random_6x3.T
print("4. Original 6x3 Array:")
print(random_6x3)
print("Transposed 3x6 Array:")
print(transposed_6x3)
print()

# 1.5 Create two random numpy arrays and compute dot product
n, p, m = 3, 4, 2
array_np = np.random.rand(n, p)
array_pm = np.random.rand(p, m)
dot_product = np.dot(array_np, array_pm)
print("5. First Array (3x4):")
print(array_np)
print("Second Array (4x2):")
print(array_pm)
print("Dot Product Result (3x2):")
print(dot_product)
```
